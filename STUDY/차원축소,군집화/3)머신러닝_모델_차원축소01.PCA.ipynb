### PCA
#### iris

- from sklearn.decomposition import PCA

```python
import pandas as pd
import numpy as np

from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
x = iris.data[:,[0,2]]
y = iris.target
```

```python
print(x.shape, y.shape)
```

```python
feature_name = [iris.feature_names[0], iris.feature_names[2]]
x_data = pd.DataFrame(x, columns=feature_name)
x_data.head()
```

```python
y_data = pd.DataFrame(y, columns = ['target'])
y_data.head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca.fit(x_data)
```

```python
pca.explained_variance_
```

```python
pca.components_
```

```python
PCscore = pca.transform(x_data)
PCscore[0:5]
```

```python
x_data[0:5]
```

```python
eigens_v = pca.components_.transpose()
print(eigens_v)
```

```python
mX = np.matrix(x)
for i in range(x.shape[1]):
    mX[:,i] = mX[:,i]-np.mean(x[:,i])
mX_df= pd.DataFrame(mX)
```

```python
(mX*eigens_v)[0:5]
```
