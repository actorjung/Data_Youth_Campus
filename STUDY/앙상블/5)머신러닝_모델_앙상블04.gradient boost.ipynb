### Gradient boost

- 참고
https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingClassifier.html

- from sklearn.ensemble import GradientBoostingClassifier

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
cancer = datasets.load_breast_cancer()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                    train_size=0.8, test_size=0.2,
                                                    random_state=156)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=10, random_state=0)

gb_clf.fit(x_train, y_train)
gb_pred = gb_clf.predict(x_test)
print(accuracy_score(y_test, gb_pred))
```

### XGboost

- 참고 https://xgboost.readthedocs.io/en/latest/python/python_intro.html

```python
import xgboost as xgb
```

```python
dtrain = xgb.DMatrix(data=x_train, label=y_train)
dtest = xgb.DMatrix(data=x_test, label=y_test)
```

```python
xgb_param = {'max_depth': 3, 'objective':'binary:logistic',
             'eval_metric':'logloss'}
```

```python
wlist = [(dtrain, 'train'), (dtest, 'eval')]

xgb_model = xgb.train(params = xgb_param, dtrain=dtrain, evals=wlist)
```

```python
xgb_model_predict = xgb_model.predict(dtest)
```

```python
xgb_model_predict
```

```python
preds = [1 if x>0.5 else 0 for x in xgb_model_predict]
preds[:5]
```

```python
print(accuracy_score(y_test, preds))
```

### LightGBM

```python
from lightgbm import LGBMClassifier
```

```python
lgbm_wrapper = LGBMClassifier()
```

```python
from lightgbm.callback import early_stopping
evals = [(x_test, y_test)]
lgbm_wrapper.fit(x_train, y_train, early_stopping_rounds=100,eval_metric='logloss',
                 eval_set=evals, verbose=True)
preds = lgbm_wrapper.predict(x_test)
```

```python
print(accuracy_score(y_test, preds))
```
