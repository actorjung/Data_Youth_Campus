### 교차분석 (Crosstab)
  - $𝑥^2$ 검정은 두가지 이상의 범주형 변수를 분석하기 위한 통계적 방법
  - 모집단에서 집단간 차이가 있는지를 분석하기 위해 사용
  - 한 변수의 범주에 따른 다른 변수의 빈도와 비율을 교차표(Cross Tabulation)로 작성하여 변수사이의 연관성을 파악

- 가설설정
  - 귀무가설(H0) : education에 따라 marital에 차이가 없다.
  - 대립가설(H1) : education에 따라 marital에 차이가 있다.

#### 데이터준비

```python
import pandas as pd

bank = pd.read_csv('bank.csv', sep=';')
bank.head()
```

```python
# bank.csv 파일을 읽어와 bank에 저장
```

#### 교차표 생성

```python
# pandas 모듈의 crosstab( ) 함수로 education과 marital 의 교차표 생성
cross_tab = pd.crosstab(bank.education, bank.marital)
cross_tab
```

```python
# margins옵션을 True로 지정하면 총합(All) 확인 가능
cross_tab = pd.crosstab(bank.education, bank.marital, margins=True)
cross_tab
```

```python
# normalize 옵션을 True로 지정하면 전체에 대한 해당 비율로 표기
cross_tab_norm = pd.crosstab(bank.education, bank.marital, margins=True, normalize=True)
cross_tab_norm
```

#### 교차분석

```python
# scipy 모듈에서 통계를 위한 stats 사용
# scipy : 과학계산을 위한 패키지로 최적화, 선형대수, 적분 등 과학계산과 관련한 함수 포함
from scipy import stats
```

```python
# stats 의 chi2_contingency( )함수를 이용해 앞서 만든 교차표의 교차분석을 실행
# 기대빈도가 array 로 출력
stats.chi2_contingency(cross_tab)
```

[결과해석]
$𝑥^2$검정결과, $𝑥^2$값이 139.09일 때 유의확률은 .000 으로서 .05에서 유의하다.
그러므로 *‘education에 따라 marital에 차이가 있다.’* 라고 할 수 있다.

### 상관분석(Correlation)
  - 상관분석은 두 연속형 변수간에 상관관계가 존재하는지를 파악하고 상관관계의 정도를 측정하는 것
  - 상관계수(Correlation Coefficient)는 데이터분석의 가장 기본이 되는 통계량으로 한 변수가 증가할 때 다른 변수의 변화와 그 정도를 추정하며, 변수들 간의 관계 정도와 방향을 나타냄
  - 주로 회귀분석을 하기 전에 독립변수↔종속변수 간의 상관이 있는지 탐색적 목적

- 가설설정
  - 귀무가설 : age와 balance 간에는 상관관계가 없다.
  - 대립가설 : age와 balance 간에는 상관관계가 있다.

```python
# stats 모듈의 f_oneway( )함수를 이용해 marital과 age 간의 평균차이분석(분산분석) 실행
# marital 범주에 따라 분류된 age 값 divorced, married, single 사용
stats.f_oneway(divorced, married, single)
```
