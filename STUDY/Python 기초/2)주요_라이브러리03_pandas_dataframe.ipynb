- #### merge() 함수

```python
fruit = pd.DataFrame({'customer_id' : [1, 3, 2, 5, 4, 3, 5, 8],
                      'fruit' : ['apple', 'banana', 'lemon', 'melon', 'peach', 'banana','peach', 'lemon'],
                      'num' : [5, 3, 10, 1, 3, 2, 7, 5]})
fruit
```

```python
customer = pd.DataFrame({'customer_id':np.arange(5),
                         'name' : ['Pengsoo', 'Pororo', 'Ryan', 'Apeach', 'Neo']})
customer
```

```python
pd.merge(customer, fruit, on='customer_id', how='inner')
```

```python
pd.merge(customer, fruit, on='customer_id', how='left')
```

```python
pd.merge(customer, fruit, on='customer_id', how='right')
```

```python
pd.merge(customer, fruit, on='customer_id', how='outer')
```

```python
cust_id = customer.set_index('customer_id')
fruits_id = fruit.set_index('customer_id')
```

```python
cust_id
```

```python
fruits_id
```

```python
pd.merge(cust_id, fruits_id, left_index = True, right_index = True)
```

- #### join() 함수 이용

```python
cust_id.join(fruits_id)
```
