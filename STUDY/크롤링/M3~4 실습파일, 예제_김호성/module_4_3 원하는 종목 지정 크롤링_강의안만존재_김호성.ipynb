### 종합

```python
import requests
import time
from fake_useragent import UserAgent
import json
from pprint import pprint
import pandas as pd

def get_symbol_code(keyword):
    """
    종목 코드 가져오는 코드
    """
    search_url = 'http://finance.daum.net/api/search?q='
    e =requests.session()
    ua = UserAgent()
    headers = {'User-Agent': ua.ie,'referer': 'http://finance.daum.net/domestic/chart'}
    res = e.get(search_url+keyword, headers=headers)
    data = res.json().get('suggestItems')[0].get('symbolCode')
    return data

def _get_news_info(page,symbol_code):
    url=f'http://finance.daum.net/content/news?page={page}&perPage=20&category=economy&searchType=all&keyword={symbol_code}&pagination=true'
    e =requests.session()
    ua = UserAgent()
    headers = {'User-Agent': ua.ie,'referer': f'http://finance.daum.net/quotes/{symbol_code}'}
    res = e.get(url, headers=headers)
    return res.json()['data']

def _get_news_contents(news_id,symbol_code):
    url=f'http://finance.daum.net/content/news/{news_id}'
    e =requests.session()
    ua = UserAgent()
    headers = {'User-Agent': ua.ie,'referer': f'http://finance.daum.net/quotes/{symbol_code}'}
    res = e.get(url, headers=headers)
    return res.json()

def _new_content_parser(html_text):
    full_text = ''
    soup = BeautifulSoup(html_text,'html.parser').select('p')
    for s in soup:
        full_text += s.text
    return full_text

def main(keyword:str, page_range:int):
    result_dict = {}
    result_list = []
    index = 1
    symbol_code = get_symbol_code(keyword) #종목코드 불러오기
    for page in range(1,page_range+1):    
        _news_info = _get_news_info(page,symbol_code) # 1페이지에 20개
        for info in _news_info:
            _data_set ={}
            news_id = info.get('newsId') #new id 
            new_data = _get_news_contents(news_id,symbol_code)
            html_text = new_data.get('content')
            title = new_data.get('title')
            date = new_data.get('createdAt')
            content=_new_content_parser(html_text) #본문파싱
            
            _data_set['index'] = index
            _data_set['title'] = title
            _data_set['contents'] = content
            _data_set['date'] = date
            
            result_list.append(_data_set)
            index +=1
            time.sleep(0.5)
            print(title)
        time.sleep(1)
        
    result_dict.update({'data':result_list})
    return result_dict
```
