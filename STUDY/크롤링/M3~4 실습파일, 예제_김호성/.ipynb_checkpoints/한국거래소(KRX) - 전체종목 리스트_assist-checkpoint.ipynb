* User-Agent : 클라이언트 소프트웨어를 식별할 수 있는 헤더 값(예: 웹 브라우저 이름과 버전)

```python
head = {'Host': 'marketdata.krx.co.kr',
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'
}
```

```python
otp_url = "http://marketdata.krx.co.kr/contents/COM/GenerateOTP.jspx"
data = {'bld': 'COM/finder_stkisu',
        'name': 'form'}
```

```python
import pandas as pd
_js = json.loads(post_code.text)['block1']
df = pd.DataFrame(_js)

lotte_df = df[df['codeName'].str.contains('롯데')]
lotte_list = lotte_df['short_code']

_df = pd.DataFrame()
i = 0
for code in lotte_list:
    for cnt in range(1,11):
        code_name = list(lotte_df['codeName'])[i]
        code = code.replace('A','')
        url = f'https://finance.naver.com/item/sise_day.nhn?code={code}&page={cnt}'
        print(url)
        df = pd.read_html(url)[0].dropna()
        df['code'] = code
        df['codeName'] = code_name
        _df = _df.append(df, ignore_index = True)
    i += 1
```

```python
_df
```

```python
rt_list = []
for cnt in range(1,16):
    payload['consonant'] = cnt
    rt3 =  requests.post("http://marketdata.krx.co.kr/contents/MKD/99/MKD99000001.jspx",  data=payload, headers=head)
    print (rt3.text)
    rt_list.append(rt3.text)
```

## json 파싱
* 결과 값이 dict 형태로 보이지만 문자열이기 때문에 json의 loads메소드를 사용하여 dict 형태로 변경합니다.
* total_df라는 빈 dataframe을 생성하고 concat 함수를 사용하여 계속 row 기준으로 dataframe을 생성하여 붙입니다.

```python
total_df = pd.DataFrame()
```

```python
total_df
```

```python
type(json.loads(rt_list[-1]))
```

```python
type(rt_list[-1])
```

```python
 pd.DataFrame(json.loads(rt_list[-1])['block1'])
```

```python
for x in rt_list:
    rt = json.loads(x)
    # 데이터프레임 생성 후 concat로 total_df로 붙입니다. 
    total_df = pd.concat([total_df, pd.DataFrame(rt['block1'])])
```
