```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
import re
```

* dart api 주소를 api_url에 저장합니다.

```python
api_url = "http://dart.fss.or.kr/api/search.json?"
```

* 삼성전자 코드를 예시를 만들어보겠습니다.
* 삼성전자의 코드는 005930입니다.

```python
value = ['005930']
```

* 전달할 데이터를 dict 형태로 만듭니다. 해당 키 값들은 Dart의 api 사용법에 나오는 내용중에서 필요한 부분만 넣은 것입니다.
* auth는 강사의 api 키입니다. 회원 가입하고 본인 키를 사용해주세요

```python
payload = {'auth' : '45eedf32874002b7d41884abba4ba52262d57c2d',
            'start_dt' : '20170101',
            'bsn_tp' : ['A002', 'A003', 'B001', 'D001'],
            'crp_cd' : value}
```

```python
r = requests.get(api_url, params=payload)
```

* params으로 전달된 dict 값으로 아래 주소가 생성됩니다. 결국 아래 주소를 실행하는 것입니다.

```python
with open("./code.csv", encoding='utf8') as f:
    company = f.read().split("\n")
```

* company 리스트에는 각 회사에 대한 정보가 하나의 문자열로 존재하고 그 값을 ","로 split() 메소드로 분리합니다. 
* company 리스트의 첫번째와 마지막 데이터를 확인해보면 사용하지 않을 데이터를 가지고 있습니다. 
* 종목 코드는 문자열의 맨 마지막에 위치하고 있습니다. split() 함수를 이용하여 콤마(,)를 기준으로 분리하여 맨 마지막을 선택합니다.
* 종목코드를 보면 Axxxxxxx 형식으로 되어 있습니다. A는 제외하고 출력할수 있게 인덱싱을 걸어줍니다.

```python
company[1].split(",")[-1][1:]
```

```python
for idx, cnt in enumerate(company[1:-1]):
    print (idx, cnt.split(",")[-1][1:])
```

* 첫째행과 마지막행은 우리가 원하는 정보가 아니기에 [1:-1] 두번째부터 마지막전까지 인덱싱을 합니다.
* for 루프를 돌면서 회사 코드를 payload안에 입력하고 api를 실행하여 각 회사별 원하는 문서 정보를 리턴받습니다.
* 리턴 받은 정보를 json.loads를 사용하여 dict로 만들고 다시 DataFrame으로 최종 생성합니다. 
* 생성된 DataFrame를 total_df라는 DataFrame과 concat() 메소드를 사용하여 누적하여 합칩니다.

```python
total_df = pd.DataFrame()
for idx, cnt in enumerate(company[1:-1]):
    # 데이터가 너무 많으니 100개만 확인해보겠습니다.
    if(idx == 100):
        print (idx)
        break
    value = cnt.split(",")[-1][1:]
    payload = {'auth' : '45eedf32874002b7d41884abba4ba52262d57c2d',
            'start_dt' : '20190101',
            'bsn_tp' : ['A003','I001'],
            'crp_cd' : value} # 회사명은 동적으로 루프를 돌면서 계속 바뀝니다. 
    r = requests.get(api_url, params=payload)
    # total_df에 전달받은 내용을 df로 만들고 합칩니다.
    total_df = pd.concat([total_df, pd.DataFrame(json.loads(r.text)['list']) ])
```

# 2.전자공시시스템에서  문서 저장하기

* 파일을 다운로드 하는 download라는 사용자 함수를 생성하겠습니다.
* 3개의 파라미터를 전달받습니다. url 주소, get 방식으로 접속할 때 사용할 데이터, 저장할 파일 이름 순으로 전달받습니다.
* Write Binary 형태로 파일을 열고 content의 값을 저정합니다.

```python
def download(url, para, filename):
    with open(filename, "wb") as f:
        res = requests.get(url, params=para)
        f.write(res.content)
```

* 파일을 다운로드 받을수 있는 url, 팝업 url, 기본 정보를 가지고 있는 url 3개를 변수에 저장합니다.

## urllib의 라이브러리의 기능을 사용하여 url 분해하기
* urllib는 python의 Standard Library입니다.
* url 주소에서 get() 방식으로 전달하는 매개변수의 값을 쉽게 분해하는 방법에 대해서 알아보겠습니다.

```python
from urllib.parse import parse_qs, urlparse
```

* 아래 예제는 html으로 제공되는 정보입니다. 이 정보의 주소를 분해해보겠습니다.
    
    > 해당 주소는 크롬의 Network 탭의 정보에서 html 정보를 가져오는 주소를 가져온것입니다.
    
* 이때 사용하는 함수는 urlparse() 입니다. 매개변수로 주소를 전달하겠습니다.

```python
# html 형태로 보여지는 문서는 아래 형식으로 보여집니다.
result = urlparse('http://dart.fss.or.kr/report/viewer.do?rcpNo=20190708800161&dcmNo=6799491&eleId=0&offset=0&length=0&dtd=HTML')
```

```python
result
```

* 이 주소를 바탕으로 html의 기본이 되는 주소를 생성합니다.

```python
html_url = result.scheme + "//" + result.netloc + result.path + "?"
print (html_url)
```

* 해당 결과에서 query 부분만 출력하겠습니다.

```python
result.query
```

* 이 값을 html이라고 따로 저장하는데 dict형태로 변경하여 저장하겠습니다.
* 이때 사용하는 함수는 parse_qs입니다. 

```python
html = parse_qs(result.query)
```

```python
html
```

* 주소를 분해하여 데이터를 구분한 이유는 형식을 가져오기 위해서 분해를 한것입니다. 
* rcpNO, dcmNO는 우리가 보고 싶은 문서 번호와 다운로드 번호이기 때문에 이 값을 해당 키의값으로 전달합니다.

```python
html['rcpNo'] = pdf_dict['rcp_no']
html['dcmNo'] = pdf_dict['dcm_no']
```

## 이제 for루프로 자동화를 합시다.
* 위에 설명했던 소스 코드를 정리하여 필요한 내용을 간추립니다.

```python
import time
```

```python
pdf_url = 'http://dart.fss.or.kr/pdf/download/pdf.do?'
popup_url = "http://dart.fss.or.kr/pdf/download/main.do?"
main_url = 'http://dart.fss.or.kr/dsaf001/main.do?'
html = {'eleId': ['0'],
 'offset': ['0'],
 'length': ['0'],
 'dtd': ['HTML']}
```

* 계속하여 url에 접속하다보면 TimeoutError 가 발생할 수 있는데 이럴때를 sleep(3) 3초 정지하였다가 다시 시도합니다. 
* 저정은 ./Dart에 저장하기 때문에 해당 폴더를 생성하거나 변경해주세요 

```python
for content in total_df.iterrows():
    print (content[1]['crp_nm'])
    url_esult = requests.get(main_url, params={'rcpNo' : content[1]['rcp_no']})
    pdf_download_re = re.findall(r"'(.*?)'",BeautifulSoup(url_esult.text, "html.parser").find("a", href="#download")['onclick'])[1]
    pdf_dict = {'rcp_no' : content[1]['rcp_no'], 'dcm_no' : pdf_download_re}
    
    # timeoutError일때 
    while(True):
        try:
            # 정상이면 break
            r2 = requests.get(popup_url, params=pdf_dict)
            break
        except 'TimeoutError':
            # error가 발생했다면... 3초 쉬었다가...
            print ('TimeoutError')
            time.sleep(3)
            pass
        
    html['rcpNo'] = pdf_dict['rcp_no']
    html['dcmNo'] = pdf_dict['dcm_no']
    if(BeautifulSoup(r2.text, 'html.parser').find("a")['href'].find("zip") > 0):
        download("http://dart.fss.or.kr/report/viewer.do?", html, "./Dart/" + content[1]['crp_nm'] + "_" + content[1]['rcp_dt'] +
                "_" + content[1]['rpt_nm'].replace("/", "").replace(":", "") + ".html")
    else:
        download(pdf_url, pdf_dict,  "./Dart/" + content[1]['crp_nm'] + "_" + content[1]['rcp_dt'] +
                "_" + content[1]['rpt_nm'].replace("/", "").replace(":", "") + ".pdf")
    
```
