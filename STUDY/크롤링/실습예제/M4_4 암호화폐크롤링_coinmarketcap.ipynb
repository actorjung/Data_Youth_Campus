#### 라이브러리 Import

```python
import requests
from bs4 import BeautifulSoup
import datetime
```

### 크롤링 사이트

```python
url = 'https://coinmarketcap.com/'
```

### 세션 연결

```python
session = requests.session()
session
```

### 응답받기

```python
response = session.get(url)
response
```

### html을 텍스트로 parsing

```python
html = response.text
```

```python
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

### result에 데이터 담기

```python
result = []
rank = soup.select('.cmc-table__cell--sort-by__rank > div')
name = soup.select('.cmc-table__cell--sort-by__name > div > a')
market_cap = soup.select('.cmc-table__cell--sort-by__market-cap > div')
price = soup.select('.cmc-table__cell--sort-by__price > a')
volume = soup.select('.cmc-table__cell--sort-by__volume-24-h > a')
supply = soup.select('.cmc-table__cell--sort-by__circulating-supply > div')
change = soup.select('.cmc-table__cell--sort-by__percent-change-24-h > div')

for i in zip(rank, name, market_cap, price, volume, supply, change):
    result.append(
        {
            'rank': i[0].text,
            'name': i[1].text,
            'market_cap': i[2].text,
            'price': i[3].text,
            'volume': i[4].text,
            'supply': i[5].text,
            'change': i[6].text
        }
    )

```

```python
result
```

### 엑셀에 데이터 담기

```python
import pandas as pd
```

```python
df = pd.DataFrame(result)
df = df.set_index('rank')
df
```

```python
df.to_excel('coinmarketcap_100.xlsx')
```

```python
result = []
date = soup.select('.cmc-table__cell--left > div')
open = soup.select('td:nth-child(2) > div')
high = soup.select('td:nth-child(3) > div')
low = soup.select('td:nth-child(4) > div')
close = soup.select('td:nth-child(5) > div')
volume = soup.select('td:nth-child(6) > div')
market_cap = soup.select('td:nth-child(7) > div')

for i in zip(date, open, high, low, close, volume, market_cap):
    result.append(
        {
            'date': i[0].text,
            'open': i[1].text,
            'high': i[2].text,
            'low': i[3].text,
            'close': i[4].text,
            'volume': i[5].text,
            'market_cap': i[6].text
        }
    )
```
