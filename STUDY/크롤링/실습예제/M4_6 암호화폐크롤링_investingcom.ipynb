### result에 Data 담기

```python
result = []
title = soup.select('#leftColumn > div.largeTitle > article > div.textDiv > a')
date = soup.select('#leftColumn > div.largeTitle > article > div.textDiv > span > span.date')
uri = soup.select('#leftColumn > div.largeTitle > article > div.textDiv > a')
summary = soup.select ('#leftColumn > div.largeTitle > article> div.textDiv > p')
for i in zip(title, date, uri, summary):
    result.append(
        {
            'title': i[0].text,
            'date': i[1].get_text().strip().replace('-\xa0', ''),
            'url': base_url + i[2].attrs['href'],
            'summary': i[3].text.strip()
        }
    )
```
