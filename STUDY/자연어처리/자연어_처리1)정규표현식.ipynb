## 정규표현식
  - 규칙을 표현하는 언어
  - 서식이 다르게 입력된 데이터에서 필요한 항목만을 확인, 추출, 교체하기 위해 사용
  - 정규표현식의 도움없이 패턴을 찾는 것(rule 기반)은 불완전
  - 파이썬에서는 정규 표현식 모듈 re을 지원하므로, 이를 이용하면 특정 규칙이 있는 텍스트 데이터를 빠르게 정제할 수 있음

- #### raw string
  - 문자열 앞에 r이 붙으면 해당 문자열이 구성된 그대로 문자열로 인식

````python
# abc\n 을 print() 함수로 출력
# \n로 줄바꿈 출력
a = 'abc\n'
print(a)
````

````python
import re
````

- #### re.compile() 함수
  - 정규표현식을 컴파일하는 함수로 파이썬에게 전해주는 역할을 함
  - 동일한 정규표현식을 다시 쓰는 번거로움을 해결

- #### re.search() 함수
  - 문자열 전체에 대해서 정규표현식과 매치되는지를 검색
  - 첫번째로 패턴을 찾으면 매치 객체 반환, 찾지 못하면 None 반환




````python
# search() 함수로 "abc"패턴이 문자열 "123abcdef"에 매치되는지 검색하여 r 변수에 저장
# 3,4,5 위치에 abc 매치
r = re.search(r'abc', '123abcdef')
print(r)
````

````python
# compile() 함수로 "abc" 패턴을 저장해 r 변수에 저장
# r 변수에 저장된 패턴을 문자열 "123abcdef"에 매치되는지 검색
# 3,4,5 위치에 abc 매치
r = re.compile('abc')
r.search('123abcdef')
````

````python
# search() 함수로 "abc"패턴이 문자열 "123abdef"에 매치되는지 검색하여 r 변수에 저장
# 매치 되는 패턴이 없어 None 반환
r = re.search(r'abc', '123abdef')
print(r)
````

````python
# compile() 함수로 "abc" 패턴을 저장해 r 변수에 저장
# r 변수에 저장된 패턴을 문자열 "123abdef"에 매치되는지 검색
# 매치되는 패턴이 없어 결과가 출력되지 않음
# print() 함수로 결과를 출력하면 None 반환
r = re.compile('abc')
r.search('123abdef')
````

````python
# search() 함수로 "\d\d\d\w"패턴이 문자열 "112abcdef119"에 매치되는지 검색하여 r 변수에 저장
# \d : 모든 숫자를 의미. [0-9]와 동일한 의미
# \w : 문자 또는 숫자를 의미. [a-zA-Z0-9]와 동일한 의미
# \d\d\d\w : 연속된 3개의 숫자와 1개의 숫자 또는 문자(숫자숫자숫자문자 또는 숫자숫자숫자숫자)
# 0,1,2,3 위치에 112a 매치
r = re.search(r'\d\d\d\w', '112abcdef119')
print(r)
````

````python
# compile() 함수로 "\d\d\d\w"패턴을 저장해 r 변수에 저장
# r 변수에 저장된 패턴을 문자열 "112abcde"에 매치되는지 검색
# 0,1,2,3 위치에 112a 매치
r = re.compile('\d\d\d\w')
r.search('112abcde')
````

### 정규 표현식 문법

- ##### 정규표현식 . 기호
  - .기호 : .은 한 개의 임의의 문자를 나타냄
    - 정규 표현식이 a.c라고 하면,  a와 c 사이에는 어떤 1개의 문자라도 올 수 있음
      - 즉, akc, azc, avc, a5c, a!c와 같은 형태는 모두 a.c의 정규 표현식과 매치됨

````python
# compile() 함수로 "a.c"패턴을 저장해 r 변수에 저장
# r 변수에 저장된 패턴을 문자열 "kkk"에 매치되는지 검색
# 매치되는 패턴이 없어 결과가 출력되지 않음
# print() 함수로 결과를 출력하면 None 반환
r = re.compile('a.c')
r.search('kkk')
````

````python
# r 변수에 저장된 패턴을 문자열 "abc"에 매치되는지 검색
# 0,1,2 위치에 abc 매치
r.search('abc')
````

- ##### 정규표현식 ? 기호
  - ?기호 : 앞의 문자가 존재할 수도 있고, 존재하지 않을 수도 있음 (문자가 0개 또는 1개)
    - 정규 표현식이 ab?c라고 한다면,
      - 이 정규 표현식에서의 b는 있다고 취급할 수도 있고, 없다고 취급할 수도 있음

````python
# compile() 함수로 "ab?c"패턴을 저장해 r 변수에 저장
r = re.compile('ab?c')
````

````python
# b가 있는 것으로 판단하여 0,1,2 위치에 abc 매치
r.search('abc')
````

````python
# b가 없는 것으로 판단하여 0,1 위치에 ac 매치
r.search('ac')
````

````python
# b가 있지만 2개이므로 패턴이 맞지 않아 None 반환
r.search('abbc')
````

- ##### 정규표현식 * 기호
  - *기호 : 앞의 문자가 무한개로 존재할 수도 있고, 존재하지 않을 수도 있음 (문자가 0개 이상)
    - 정규 표현식이 ab*c라고 한다면,
      - ac, abc, abbc, abbbc 등과 매치할 수 있으며 b의 개수는 무수히 많아도 됨

````python
# compile() 함수로 "ab*c"패턴을 저장해 r 변수에 저장
r = re.compile('ab*c')
````

````python
# b가 없는 것으로 판단하여 0,1 위치에 ac 매치
r.search('ac')
````

````python
# b가 있는 것으로 판단하여 0,1,2 위치에 abc 매치
r.search('abc')
````

````python
# b가 있는 것으로 판단하여 0,1,2,3,4,5,6 위치에 abbbbbc 매치
r.search('abbbbbbbc')
````

````python
# b가 없으나 c가 없어 패턴이 맞지 않아 None 반환
r.search('a')
````

- ##### 정규표현식 + 기호
  - +기호 : 앞의 문자가 최소 한 개 이상 존재 (문자가 1개 이상)
    - +는 *와 유사하지만, 앞의 문자가 최소 1개 이상이어야 한다는 점이 다름
    - 정규 표현식이 ab+c라고 하면,
      - ac는 매치되지 않지만 abc, abbc, abbbc 등과 매치할 수 있으며 b의 개수는 무수히 많을 수 있음

````python
# compile() 함수로 "ab+c"패턴을 저장해 r 변수에 저장
r = re.compile('ab+c')
````

````python
# b가 없는 것으로 판단하여 None 반환
r.search('ac')
````

````python
# b가 있는 것으로 판단하여 0,1,2 위치에 abc 매치
r.search('abc')
````

````python
# b가 있는 것으로 판단하여 0,1,2,3,4,5,6 위치에 abbbbbc 매치
r.search('abbbbbbbc')
````

- ##### 정규표현식 ^ 기호
  - ^기호 : ^는 시작되는 글자를 지정
    - 정규표현식이 ^a라면, a로 시작되는 문자열만을 찾아냄

````python
# compile() 함수로 "^a"패턴을 저장해 r 변수에 저장
r = re.compile('^a')
````

````python
# ab는 a로 시작되기 때문에 0위치에 a 매치
r.search('ab')
````

````python
# bcad는 a로 시작되지 않기 때문에 None 반환
r.search('bcad')
````

- ##### 정규표현식 \$ 기호
  - \$기호 : \$는 끝나는 글자를 지정
    - 정규표현식이 a\$라면, a로 끝나는 문자열만을 찾아냄


````python
# compile() 함수로 "a$"패턴을 저장해 r 변수에 저장
r = re.compile('a$')
````

````python
# cba는 a로 끝나기 때문에 2위치에 a 매치
r.search('cba')
````

````python
# a는 a로 끝나기 때문에 0위치에 a 매치
r.search('a')
````

````python
# abc는 a로 끝나지 않기 때문에 None 반환
r.search('abc')
````

- ##### 정규표현식 {숫자} 기호
  - {숫자} 기호 : 문자에 해당 기호를 붙이면, 해당 문자를 숫자만큼 반복
    - 정규 표현식이 ab{2}c라면,
      - a와 c 사이에 b가 존재하면서 b가 2개인 문자열에 대해서 매치

````python
# compile() 함수로 "ab{2}c"패턴을 저장해 r 변수에 저장
r = re.compile('ab{2}c')
````

````python
# abc는 a와 c사이에 b가 존재하지만 1개이므로 패턴이 맞지 않아 None 반환
r.search('abc')
````

````python
# abbc는 a와 c사이에 b가 2개 존재하기 때문에 0,1,2,3위치에 abbc 매치
r.search('abbc')
````

````python
# abbbbbc는 a와 c사이에 b가 존재하지만 패턴이 맞지 않아 None 반환
r.search('abbbbbc')
````

- ##### 정규표현식 {숫자1,숫자2} 기호
  - {숫자1,숫자2} 기호 : 문자에 해당 기호를 붙이면, 문자를 숫자1 이상 숫자2 이하만큼 반복
    - 정규 표현식이 ab{2,8}c라면,
      - a와 c 사이에 b가 존재하면서 b는 2개 이상 8개 이하인 문자열에 대해서 매치

````python
# compile() 함수로 "ab{2,8}c"패턴을 저장해 r 변수에 저장
r = re.compile('ab{2,8}c')
````

````python
# abc는 a와 c사이에 b가 존재하지만 1개이므로 패턴이 맞지 않아 None 반환
r.search('abc')
````

````python
# abbc는 a와 c사이에 b가 2개 존재하기 때문에 0,1,2,3위치에 abbc 매치
r.search('abbc')
````

````python
# abbbbbc는 a와 c사이에 b가 2개 이상 8개 이하로 존재하기 때문에 0,1,2,3,4,5,6위치에 abbbbbc 매치
r.search('abbbbbc')
````

````python
# abbbbbbbbbbbbbbbbbbbbbbc는 a와 c사이에 8개 초과로 존재하기 때문에 패턴이 맞지 않아 None 반환
r.search('abbbbbbbbbbbbbbbbbc')
````

- ##### 정규표현식 {숫자,} 기호
  -  {숫자,}  기호 : 문자에 해당 기호를 붙이면 해당 문자를 숫자 이상 만큼 반복
    - 정규 표현식이 a{2,}bc라면,
      - 뒤에 bc가 붙으면서 a의 갯수가 2개 이상인 경우인 문자열과 매치
    - 만약 {0,}을 쓴다면 *와 동일한 의미가 되며, {1,}을 쓴다면 +와 동일한 의미가 됨


````python
# compile() 함수로 "ab{2,}c"패턴을 저장해 r 변수에 저장
r = re.compile('ab{2,}c')
````

````python
# abc는 a와 c사이에 b가 존재하지만 패턴이 맞지 않아 None 반환
r.search('abc')
````

````python
# abbc는 a와 c사이에 b가 2개 존재하기 때문에 0,1,2,3위치에 abbc 매치
r.search('abbc')
````

````python
# abbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbbc는 a와 c사이에 b가 2개 이상 존재하기 때문에 매치
r.search('abbbbbbbbbbbbc')
````

- ##### 정규표현식 [문자] 기호
  - \[ ] 기호 : [ ]안에 문자들을 넣으면, 그 문자들 중 한 개의 문자와 매치라는 의미
    - 정규 표현식이 [abc]라면,
      - a 또는 b또는 c가 들어가있는 문자열과 매치됨
    - [a-zA-Z]는 모든 알파벳,  [0-9]는 모든 숫자 와 같이 범위를 지정하는 것도 가능

````python
# compile() 함수로 "[abc]"패턴을 저장해 r 변수에 저장
r = re.compile('[abc]')
````

````python
# aaa는 a가 존재하기 때문에 0위치에 a 매치
r.search('aaa')
````

````python
# abbbc는 a가 존재하기 때문에 0위치에 a 매치
r.search('abbbc')
````

````python
# cba는 c가 존재하기 때문에 0위치에 c 매치
r.search('cba')
````

````python
# ddd는 a도, b도, c도 존재하지 않기 때문에 패턴이 맞지 않아 None 반환
r.search('ddd')
````

````python
# dcf는 c가 존재하기 때문에 1위치에 c 매치
r.search('dcf')
````

- ###### 정규표현식을 알파벳 소문자로 지정

````python
# compile() 함수로 "[a-z]"패턴을 저장해 r 변수에 저장
# 정규표현식을 알파벳 소문자로 지정
r = re.compile('[a-z]')
````

````python
# Pengsoo는 알파벳 소문자 e가 존재하므로 1위치에 e 매치
r.search('Pengsoo')
````

````python
# 111은 알파벳 소문자가 존재하지 않으므로 None 반환
r.search('111')
````

- ##### 정규표현식 [^문자] 기호
  - [^문자] 기호 : ^ 기호 뒤에 붙은 문자들을 제외한 모든 문자를 매치
    - 정규표현식이 [^abc] 라면,
      - a 또는 b 또는 c가 들어간 문자열을 제외한 모든 문자열을 매치함

````python
# compile() 함수로 "[^abc]"패턴을 저장해 r 변수에 저장
# [^abc] : a 또는 b 또는 c가 들어간 문자열을 제외한 모든 문자열을 매치함
r = re.compile('[^abc]')
````

````python
# a는 a가 존재하므로 None 반환
r.search('a')
````

````python
# ab는 a가 존재하므로 None 반환
r.search('ab')
````

````python
# pengsoo는 a도, b도, c도 존재하지 않으므로 0위치에 p 매치
r.search('pengsoo')
````

````python
# 111은 a도, b도, c도 존재하지 않으므로 0위치에 1 매치
r.search('111')
````

- ##### 정규표현식 | 기호
  - | 기호 : AlB와 같이 쓰이며 A 또는 B의 의미
  - 정규표현식이 [a|A] 라면,
    - a 또는 A 가 들어간 문자열을 매치함

````python
# compile() 함수로 "[^abc]"패턴을 저장해 r 변수에 저장
# [a|A] : a 또는 A 가 들어간 문자열을 매치함
r = re.compile('[a|A]')
````

````python
# banana는 a가 존재하므로 1위치에 a 매치
r.search('banana')
````

````python
# BANANA는 A가 존재하므로 1위치에 A 매치
r.search('BANANA')
````

````python
# pengsoo는 a도 A도 존재하지 않으므로 None 반환
r.search('pengsoo')
````

### 정규표현식 함수

- ##### re.match() 함수
  - search() 와 비슷하나, 문자열의 처음부터 비교하여 정규표현식과 매치되는지를 검색
  - 시작에 해당 패턴이 존재하지 않으면 None 반환

````python
# compile() 함수로 "ab."패턴을 저장해 r 변수에 저장
# ab. : a와 b 다음 어떤 1개의 문자
r = re.compile('ab.')
````

````python
# search() 함수로 "ab."패턴이 문자열 "kkkabc"에 매치되는지 검색
# 3,4,5 위치에 abc 매치
r.search('kkkabc')
````

````python
# match() 함수로 "ab."패턴이 문자열 "kkkabc"에 처음 문자열부터 매치되는지 검색
# 시작에 ab. 패턴이 존재하지 않으므로 None 반환
r.match('kkkabc')
````

````python
# match() 함수로 "ab."패턴이 문자열 "abckkk"에 처음 문자열부터 매치되는지 검색
# 0,1,2 위치에 abc 매치
r.match('abckkk')
````

````python
# compile() 함수로 "\d\d\d"패턴을 저장해 r 변수에 저장
# \d : 모든 숫자를 의미. [0-9]와 동일한 의미
# \d\d\d : 연속된 3개의 숫자
r = re.compile('\d\d\d')
````

- ##### re.sub() 함수
  - 문자열에서 정규표현식과 일치하는 부분에 대해서 다른 문자열로 대체(replace)하여 결과를 문자열로 다시 반환
  - 대체할 두번째 인자는 특정 문자열이 될 수도 있고, 함수가 될 수도 있음
  - 자연어 처리를 위해 특수 문자를 제거하고 싶다면 알파벳 외의 문자는 공백으로 처리하는 등의 사용 용도로 쓸 수 있음



```
txt = "Name : Pengsoo, Phone : 010 - 1234 - 1111, Age : 10"
```



````python
# txt 문자열 선언
txt = "Name : Pengsoo, Phone : 010 - 1234 - 1111, Age : 10"
````

````python
# sub() 함수로 txt 에서 알파벳 소문자, 대문자가 아닌 경우 공백으로 대체
# 특수문자, 숫자 등이 공백으로 대체됨
print(re.sub('[^a-zA-Z]', ' ', txt))
````

````python
# sub() 함수로 txt 에서 알파벳 소문자, 대문자가 아닌 경우 제거
# 특수문자, 숫자 등이 모두 삭제됨
print(re.sub('[^a-zA-Z]', '', txt))
````

````python
# sub() 함수로 txt 에서 알파벳 소문자, 대문자가 아닌 경우 * 로 대체
# 특수문자, 숫자 등이 * 로 대체됨
print(re.sub('[^a-zA-Z]', '*', txt))
````

### grouping
  - 매칭 결과를 각 그룹별로 분리할 수 있음
  - 패턴을 명시할 때 분리할 그룹을 괄호() 로 묶어주면 됨

````python
# compile() 함수로 "(\w+)@(.+)"패턴을 저장해 r 변수에 저장
# \w : 문자 또는 숫자를 의미. [a-zA-Z0-9]와 동일한 의미
# . : 한 개의 임의의 문자
# (\w+)@(.+) : (문자가 1개 이상)@(임의의 문자가 1개 이상)
r = re.compile("(\w+)@(.+)")
````

````python
# search() 함수로 "(\w+)@(.+)"패턴이 문자열 "test@abc.com"에 매치되는지 검색하여 r_gp 변수에 저장
# group() 함수로 매칭 결과를 분리하여 확인
# 0 : 그룹별 불리 없이 매칭결과 확인
# 1 : 첫번째 그룹만 확인 (괄호 순서대로)
# 2 : 두번째 그룹만 확인 (괄호 순서대로)
r_gp = r.search('test@abc.com')
print(r_gp.group(1))
print(r_gp.group(2))
print(r_gp.group(0))
````
