```python
# SQLite DB 연결
conn = sqlite3.connect("c:/python_class/hello_db.db")
 
# Connection 으로부터 Cursor 생성 커서(Cursor)는 일련의 데이터에 순차적으로 액세스할 때 검색 및 "현재 위치"를 포함하는 데이터 요소
cur = conn.cursor()

# query = """
# select strftime('%Y-%m-%d %H:%M', t_20.reg_date) as grp_time, sum(replace(substr(t_20.market_cap,2), ',', '')) as sum_market_cap_top_20 
# from (
# select * from coinmarketcap_new where rank <= 20
# ) t_20
# group by grp_time
# """

# 데이터 넣기
query = """
INSERT INTO top_20_volume (grp_time, sum_market_cap_top_20)
select strftime('%Y-%m-%d %H:%M', t_20.reg_date) as grp_time, sum(replace(substr(t_20.market_cap,2), ',', '')) as sum_market_cap_top_20 
from (
select * from coinmarketcap_new where rank <= 20
) t_20
group by grp_time
"""

# SQL 쿼리 실행
cur.execute(query)

conn.commit()
 
# 데이타 Fetch(가져옴)
rows = cur.fetchall()
for row in rows:
    print(row)
 
# Connection 닫기
conn.close()
```

### 데이터프레임 활용

```python
import pandas as pd
```

```python
conn = sqlite3.connect("c:/python_class/hello_db.db")

# 데이터 추출 Top20에 대한 Volume 합
query = """
select strftime('%Y-%m-%d %H:%M', t_20.reg_date) as grp_time, sum(replace(substr(t_20.market_cap,2), ',', '')) as sum_market_cap_top_20 
from (
select * from coinmarketcap_new where rank <= 20
) t_20
group by grp_time
"""


# 쿼리 읽어서 Dataframe으로 만들기
df = pd.read_sql(query, conn)

# 데이터프레임으로 테이블생성해서 넣기
df.to_sql('top_20_volume', conn, if_exists='append', index=False)

conn.close()
```
